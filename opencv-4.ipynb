```python
#importing libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading the image
image=cv2.imread("images\lion.jpg")
```

```python
#displaying the image
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
lion=plt.imshow(image)
print(lion)
```

```python
#removing the noise in an image
deNoising=cv2.fastNlMeansDenoisingColored(image, None, 40, 40, 14, 30)
#the last numbers are the values, for each image they are different so you'll have to insert different values.
```

```python
#displaying the new image
dn_lion=plt.imshow(deNoising)
print(dn_lion)
```

```python
#removing the noise in an image
deNoising=cv2.fastNlMeansDenoisingColored(image, None, 60, 60, 21, 45)
#changing the parameters
```

```python
#displaying the new image
dn_lion=plt.imshow(deNoising)
print(dn_lion)
```
