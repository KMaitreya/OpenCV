```python
#importing libraries
import cv2
import matplotlib.pyplot as plt
```

```python
#cascade classifier for face detection
face_cascade=cv2.CascadeClassifier("files/haarcascade_frontalface_default.xml")
```

```python
#reading the image
img=cv2.imread("images\person.jpg")
```

```python
#displaying the image
person=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(person)
```

```python
#cascade classifiers work on grayscale hence we will have to convert the image to grayscale
gray_person=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_person, cmap="gray")
```

```python
#detecting the face
face=face_cascade.detectMultiScale(gray_person, 1.3, 5)
#arguments                             image      scale  neighbours
#in this case we only have one face so it doesn't matter but incase we want to detect faces in a group it will matter
#play around with the scale adn neigbours parameter to get different results
#this fucntion will return us the diagonal coordinates of the face
```

```python
#printing the face diagonal coordinates
print(face)
```

```python
#we will now use these coordinates to draw the boundary rectangle
for(x, y, w, h) in face:
    cv2.rectangle(person, (x, y), (x+w, y+w), (0, 255, 0), 5)
```

```python
#displaying the image with the detected face
plt.imshow(person)
```
