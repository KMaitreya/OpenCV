```python
#importing libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the image
image=cv2.imread("images\color.jpg")
```

```python
#conversion to canny image
gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
canny=cv2.Canny(gray, 150, 200)
canny_image=plt.imshow(canny, cmap='gray')
print("Edges\n", canny_image)
```

```python
#erosion(fading the edges)
kernel=np.ones((2, 2), np.uint8)       #higher the matrix value more erosion will take place on the image
erode=cv2.erode(canny, kernel, iterations=1)
erode_image=plt.imshow(erode, cmap="gray")
print("Eroded image\n", erode_image)
```

```python
#dilation(making the edges prominant)
kernel=np.ones((3, 3), np.uint8)
dilate=cv2.dilate(canny, kernel, iterations=1)
dilate_image=plt.imshow(dilate, cmap="gray")
print("Dilated image\n", dilate_image)
```

```python
#erosion after dialtion will give us better results

#dialtion
kernel_d=np.ones((5, 5), np.uint8)
dilate=cv2.dilate(canny, kernel_d, iterations=1)

#erosion
kernel_e=np.ones((5, 5), np.uint8)
erode=cv2.erode(dilate, kernel_e, iterations=1)
imageF=plt.imshow(erode, cmap="gray")
print("Dilated and eroded image\n", imageF)
```
