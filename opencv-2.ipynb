```python
#import libraries
import cv2
import matplotlib.pyplot as plt
```

```python
#reading the image
image=cv2.imread("images\color.jpg")
```

```python
#displaying the image
bgr_image=plt.imshow(image)
print(bgr_image)
```

```python
#conversion of image to grayscale
gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#displaying edges in the image using Canny()
canny=cv2.Canny(gray, 150, 200)
canny_image=plt.imshow(canny, cmap='gray')
print("Edges\n", canny_image)
```
