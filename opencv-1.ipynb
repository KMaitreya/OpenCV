```python
#importing library
import cv2
import matplotlib.pyplot as plt
```

```python
#reading the image
image=cv2.imread(r"C:\Users\Maitreya\JupyterNotebooks\OpenCV\images\color.jpg")
```

```python
#original(BGR)
bgr_image=plt.imshow(image)
print("BGR image: ", image.shape, "\n", bgr_image)
```

```python
#RGB
rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
rgb_image=plt.imshow(rgb)
print("RGB image: ", rgb.shape, "\n", rgb_image)
```

```python
#grayscale
gray=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image=plt.imshow(gray, cmap="gray")
print("Grayscale image: ", gray.shape, "\n", gray_image)
```

```python
#HSV
hsv=cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
hsv_image=plt.imshow(hsv)
print("HSV image: ", hsv.shape, "\n", hsv_image)
```
