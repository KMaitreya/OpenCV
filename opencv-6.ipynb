```python
#importing libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading the image on which colour detection will take place
img=cv2.imread("images/shapes.png")
```

```python
#displaying the image
rgb_shapes=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)      #converting BGR image to RGB image
plt.imshow(rgb_shapes)
```

```python
#converting the bgr image to an hsv image
hsv_shapes=cv2.cvtColor(rgb_shapes, cv2.COLOR_RGB2HSV)
plt.imshow(hsv_shapes)
```

```python
#isolating images based on their hue value

#isolating the blue triangle
lower_bound=np.array([65, 0, 0])
upper_bound=np.array([110, 255, 255])
#creating a mask so that we will only be able to see the blue triangle
mask_triangle=cv2.inRange(hsv_shapes, lower_bound, upper_bound)
triangle=cv2.cvtColor(mask_triangle, cv2.COLOR_BGR2RGB)
plt.imshow(triangle)
```

```python
#displaying the isolated triangle
isolated_triangle=cv2.bitwise_and(triangle, img, mask=mask_triangle)
final_triangle=cv2.cvtColor(isolated_triangle, cv2.COLOR_BGR2RGB)
plt.imshow(final_triangle)
```
